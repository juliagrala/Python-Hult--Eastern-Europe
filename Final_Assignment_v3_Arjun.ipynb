```python
#34. Male Female Age Distribution (graph only)

# Setting the positions and width for the bars
pos = list(range(len(teamDataCleaned_df['PopulationMale']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(20,10))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        teamDataCleaned_df['PopulationMale'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=teamDataCleaned_df['CountryName'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        teamDataCleaned_df['PopulationFemale'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=teamDataCleaned_df['CountryName'][1]) 

# Set the y axis label
ax.set_ylabel('Numbers in Millions')

# Set the chart's title
ax.set_title('Male Female Age Distribution')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(teamDataCleaned_df['CountryName'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(teamDataCleaned_df['PopulationFemale']*1.1)])

# Adding the legend and showing the plot
plt.legend(['Male', 'Female'], loc='upper left')
plt.grid()
plt.show()
```

```python
#Population Density vs Surface Area (Graph Only)

# Setting the positions and width for the bars
pos = list(range(len(teamDataCleaned_df['SurfaceArea']))) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(20,10))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        teamDataCleaned_df['SurfaceArea'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=teamDataCleaned_df['CountryName'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        teamDataCleaned_df['PopulationDensity'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=teamDataCleaned_df['CountryName'][1]) 

# Set the y axis label
#ax.set_ylabel('Numbers in Millions')

# Set the chart's title
ax.set_title('Population Density vs Surface Area Distribution')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(teamDataCleaned_df['CountryName'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(teamDataCleaned_df['SurfaceArea']*1.1)])

# Adding the legend and showing the plot
plt.legend(['Surface Area', 'Population Density'], loc='upper left')
plt.grid()
plt.show()
```
